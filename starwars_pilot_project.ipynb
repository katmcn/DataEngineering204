10 or less entries per page

```python
#Create a list to store results
starships = []
```

```python
#Request the data
print("Downloading the search results")
response = requests.get("https://swapi.dev/api/starships")
data = response.json()

#Store results from initial search (from the first page)
starships = starships + data['results']

#While the page shows a next page
while data['next'] is not None:
    print("Next page found, downloading...", data['next'])
    response = requests.get(data['next'])
    data = response.json()
    starships = starships + data['results']

print("We have", len(starships), "total results")
```

```python
ships_urls = {}
for ship in starships:
    ships_urls[ship["name"]] = ship['pilots']
```

```python
ships_urls
```

```python
#This runs but only sets one pilot name per ship
ship_pilots = {}
for k, v in ships_urls.items():
    if v == []:
        continue
    else:
        for url in v:
            response = requests.get(url)
            person_url = response.json()
            ship_pilots[k] = person_url['name']

```
